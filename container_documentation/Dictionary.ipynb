**<ins>DICTIONARY</ins>**

Dictionaries are built-in data types in Python.Dictionaries are mutable, and can contain key:value pairs.

In Python 3.7 and later, dictionaries maintain the insertion order of their items, which means they are ordered.


Values in a dictionary can be of any data type and can be duplicated, whereas keys canâ€™t be repeated and must be immutable.


Dictionaries in Python differ from sequences in that they are indexed by keys instead of a range of numbers. These keys can be of any immutable type, making strings and numbers suitable candidates. Tuples can also serve as keys if they solely consist of strings, numbers, or other tuples. However, if a tuple contains any mutable object, either directly or indirectly, it cannot be utilized as a key.

Importantly, lists cannot function as keys in dictionaries due to their mutability. Lists can be modified in place using operations such as index assignments, slice assignments, and methods like append() and extend(). This mutability prevents them from serving as reliable keys for dictionary indexing.

```python
#To demonstrate that dictionaries are ordered
dicto={'a': 1, 'b': 2, 'c': 3, 'd': 4}
print("Original order of dictionary:")
for key, value in dicto.items():
    print(key, value)

#Modify the dictionary
dicto['b']=10
print("\nOrder after modification:")
for key, value in dicto.items():
    print(key, value)
```

This behavior ensures that the order in which items are inserted into the dictionary is preserved when iterating over the dictionary

```python
#Both a and b look differently ordered but when compared return True

b = {'three':3, 'one':1,"two":2}
print (b)

a = {'one':1, 'two':2, 'three':3}
print(a)
print(a==b)
```

**Dictionary Methods**

Dictionaries can be manipulated through various methods:

1) keys():

Returns a view object. The view object contains a list of the keys in the dictionary.
Syntax: dict.keys()

```python
a.keys()
```

2) values():

Returns a view object that displays a list of all the values in the dictionary. Syntax: dict.values()

```python
a.values()
```

3) items():

Returns a view object that displays a list of key-value tuples in the dictionary. Syntax: dict.items()


```python
a.items()
```

These view objects are dynamic, meaning that if the underlying dictionary changes (keys are added or removed), the view reflects those changes automatically.

However, the view objects themselves are not actual lists or sets; they are more memory-efficient, providing a dynamic and up-to-date view on the dictionary without duplicating the data.

```python
my_dict = {'one': 1, 'two': 2, 'three': 3}

# Get a view of keys
keys_view = my_dict.keys()

# Add a new key-value pair to the dictionary
my_dict['four'] = 4

# Print the view again
print(keys_view)

#keys_view will automatically include the new key 'four' because it reflects the changes made to the underlying dictionary.
```

4) clear():

Removes all the elements from the dictionary.

Syntax: dict.clear()

```python
a.clear()
print(a)
#Returns empty dictionary
```

5) get():

Returns the value of the specified key

Synatx: dict.get(key)

```python
print(b)
b.get('three','None')
#Suppose the key is not present
b.get('ten','Key not present')
#It prints the default value
```

6) pop():

Removes the element with the specified key

Syntax: dict.pop(key)

```python
b.pop('one')
print(b)
```

7) popitem():

Removes the last key-value pair that we have inserted.

Synatx: dict.popitem()

```python
b.popitem()
print(b)
```

8) setdefault():

Sets the key to a default value if the key is not present in the dictionary.

Synatx: dict.setdefault(key, default="None")

```python
b.setdefault("four",5)
print(b)
```

9) update() :     

adds a key:value pair to the dictionary

Synatx: dict.update([mapping])

```python
b.update({'six':6})
print(b)
```

10) dict() constructor:

It builds dictionaries directly from sequence of key-value pairs.

```python
d= dict([('red', 4), ('blue', 7), ('pink', 8)])
print(d)
```

11) copy():        

Returns a shallow copy of the dictionary.

Synatx:  dict.copy()

```python
c=b.copy()
print(c)
b['three']= 10
print(c)

#altering the objects in original dictionary will affect itâ€™s copy as well
```

12) fromekeys():  

Returns a new dictionary with keys from a supplied iterable and values all set to specified value

Syntax: dict.fromkeys(iterable[,value])

```python
nums=[12,13,14,15]
num_dict={}
num_dict=num_dict.fromkeys(nums)
print(num_dict)
#or
nums_dict={}.fromkeys(nums, "Numbers")
print(nums_dict)
```
