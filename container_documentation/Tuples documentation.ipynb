**INTRODUCTION**


Tuples are immutable sequences, typically used to store collections of heterogeneous data. Tuples, similar to lists, provide unique characteristics that make them valuable in specific scenarios.


1. <ins>Creation</ins>

Tuples may be constructed in a number of ways:

* Using a pair of parentheses to denote the empty tuple: `()`









```python
t1 = () # an empty tuple
print(t1)
```

* Using a trailing comma for a singleton tuple: a, or `(a,)`

```python
t2 = 1, # single element tuples have trailing commas
print(t2)
t = (1)
print(type(t))
#The above will not be recognized as a tuple
```

* Separating items with commas: a, b, c or `(a, b, c)`

```python
t3 = 1,2,3,4 # to create a tuple, you do not require parentheses, you just require commas
print(t3)
```

* Using the `tuple()` built-in: `tuple()` or `tuple(iterable)`

The constructor builds a tuple whose items *are the same and in the same order as iterableâ€™s items*. iterable may be either a sequence, a container that supports iteration, or an iterator object. If iterable is already a tuple, it is returned unchanged. For example, `tuple('abc')` returns `('a', 'b', 'c')` and` tuple( [1, 2, 3] ) `returns (1, 2, 3). If no argument is given, the constructor creates a new empty tuple, `()`.

Note that it is actually the comma which makes a tuple, not the parentheses. The parentheses are optional, except in the empty tuple case, or when they are needed to avoid syntactic ambiguity. For example, `f(a, b, c)` is a function call with three arguments, while `f((a, b, c))` is a function call with a 3-tuple as the sole argument.

```python

list1 = ['blue', 3, True]
t = tuple(list1)
print(t)

#We can also use range() to create tuples
print(tuple(range(1, 7)))
```

* Tuple with mixed data types

Tuples can store elements of different data types, including numbers, strings, and boolean values.

```python
mixed_tuple = ('apple', 3.14, 42, True)
print(mixed_tuple)
```

**IMMUTABILITY OF TUPLES**

Once a tuple is created, its elements cannot be modified, added, or removed. This immutability ensures data integrity and consistency

```python
#Tuples are immutable
#Trying to change an element raises an error for eg
t3[0]= 5
print(t3)
#As we can see, TypeError is raised
```

```python
#Tuples can contain mutable objects
#They can even be nested tuples
t1=(1,2,[1,2,3])
t2=(t1, (5,6,7))
print(t1)
print(t2)
```

`Implications of Immutability`

1) Immutability makes tuples suitable for use as keys in dictionaries, ensuring data safety.

2) Tuples are suitable for situations where data should remain constant throughout program execution.

2.<ins>Access</ins>

Accessing elements in a tuple in Python is similar to accessing elements in a list.


Tuples use zero-based indexing, meaning the first element has an index of 0, the second has an index of 1, and so on.





```python
print(t1[0], t1[1]) #indexing

```


You can also use negative indexing, where -1 refers to the last element, -2 to the second-to-last, and so forth.

```python
print(t1[-1]) #negative indexing
```

3. <ins>Methods</ins>


* `index()`

The index() method in Python is used to find the index of the first occurrence of a specified value within a tuple. If the specified element is not found, it raises a ValueError. 

`tuple.index(element, start, end)`

* `element`: The value for which the method searches.
`start (optional)`: The index from where the search begins. The default is 0.
`end (optional)`: The index at which the search stops. The default is the end of the tuple.


Source of the image: https://pythonexamples.org/python-access-tuple/

```python
my_tuple = (10, 20, 30, 20, 40, 50)

# Find the index of the first occurrence of 20
index_20 = my_tuple.index(20)
print("Index of first occurrence of 20:", index_20)  # Output: 1

```

```python
# Find the index of 20 starting from index 2
index_20_from_2 = my_tuple.index(20, 2)
print("Index of 20 starting from index 2:", index_20_from_2)  # Output: 3

```

```python
# Find the index of 40 in the subset [2:5]
index_40_in_subset = my_tuple.index(40, 2, 5)
print("Index of 40 in subset [2:5]:", index_40_in_subset)  # Output: 4
```

* `count`


The count() method in Python is used to count the number of occurrences of a specific element within a tuple. 

```python
my_tuple = (10, 20, 30, 20, 40, 50)

# Count the number of occurrences of 20
count_20 = my_tuple.count(20)
print("Number of occurrences of 20:", count_20)  

```

```python

# Count the number of occurrences of 60 (not present in the tuple)
count_60 = my_tuple.count(60)
print("Number of occurrences of 60:", count_60)  
```

**PACKING AND UNPACKING OF TUPLES**

Packing refers to combining multiple values into a single container, while unpacking involves extracting values from a container. 







Source of the image: https://prepinsta.com/python/packing-and-unpacking-in-python-arguments/

```python
#Packed tuple
#It occurs implicitly when multiple values are separated by commas.
my_tuple=1, 'blue', 5, 'a'
```

The asterisk (*) operator allows `packing` multiple values into a list or tuple or unpacking them

```python
num = 1, 2, 3, 4, 5
*front, back= num

#front is a list containing the first three values
# back is a tuple containing the last two values.

print(front)
print(back)


```

In Python, you can unpack a tuple by assigning its values to variables in a single line. 


```python

# Unpack the tuple
a, b, c = my_tuple

# Now, 'a' will be 1, 'b' will be 2, and 'c' will be 3
print(a)  
print(b)  
print(c)  

```

You can also use the `*` operator to unpack the remaining values as a list.

```python
my_tuple2 = (1, 2, 3, 4, 5)

# Unpack the first two values and collect the rest in a list
a, b, *rest = my_tuple2

print(a)    
print(b)    
print(rest)
```

**<ins>USE CASES</ins>**

* `Dictionary Packing/Unpacking`

```python
# Packing
student_info = {'name': 'Pranav', 'age': 19, 'grade': 'A'}
packed_dict = {**student_info, 'status': 'present'}

# Unpacking
name, age, grade = student_info.values()
```

* `String Formatting`

```python
# Packing
format_string = "Name: {}, Age: {}".format(*("Pranav", 19))

# Unpacking
name, age = format_string.split(", ")
```

* `Simultaneous Assignment`

```python
# Packing
a, b = 1, 2

# Unpacking
a, b = b, a  # Swap values of a and b
```
